```python
def metrics_evaluation(target , prediction) :
    print('confusion matrix \n' , confusion_matrix(target , prediction))
    print()
    print('accuracy \n' , accuracy_score(target , prediction))
    print()
    print('recall \n'   , recall_score(target , prediction))
    print()
    print('precision \n'   , precision_score(target , prediction))
    print()
    print('f1 score \n'   , f1_score(target , prediction))
```

```python
diabets.target = diabets.iloc[:,-1]
diabets.data = diabets.iloc[:, :-1]
```

```python
X_train , X_test , y_train , y_test = train_test_split(diabets.data , 
                                                       diabets.target , 
                                                       test_size = 0.2 , 
                                                       random_state = 100)

X_train.shape , X_test.shape , y_train.shape , y_test.shape 
```

```python
# 표준화 진행 

dia_model01 = LogisticRegression()

s_scale = StandardScaler()
s_scale.fit(X_train)

strain_data = s_scale.transform(X_train)
stest_data = s_scale.transform(X_test)


dia_model01.fit(strain_data, y_train)
y_pred = dia_model01.predict(stest_data)

accuracy_score(y_test, y_pred)
```

```python
# 정규화 진행 
dia_model02 = RandomForestClassifier()

m_scale = MinMaxScaler()
m_scale.fit(X_train)

mtrain_data = m_scale.transform(X_train)
mtest_data = m_scale.transform(X_test)


dia_model02.fit(mtrain_data, y_train)
y_pred = dia_model02.predict(mtest_data)

accuracy_score(y_test, y_pred)
```

```python
# 최적의 파라메터 GridSearchCV 사용 
params02 = {'max_depth' : [2,3,5,10],
         'min_samples_split' : [2,3,5],
         'min_samples_leaf':[1,5,8]}

params03 = {
    'criterion' : ['gini' , 'entropy'],
    'splitter' : ['best' , 'random'], # spllitter라고 했다. 오타 주의 
    'max_depth' : [1,2,3] , 
    'min_samples_split' : [2, 3 ]
}
params = {
    'n_estimators' : [200, 300] , 
    'max_features' : [6, 8, 15, 20] , 
    'max_depth'    : [4, 6, 8]
}

```

```python
# lr params 
param_lr = {
      'max_iter' : [10,50,100] 
         }
```

```python
gsc_model_01 = GridSearchCV(dia_model01, param_grid=param_lr, cv=10 , refit=True ,scoring='recall')
gsc_model_01.fit(X_train , y_train)
```

```python
estimator = gsc_model_01.best_estimator_
y_pred = estimator.predict(X_test)
metrics_evaluation(y_test , y_pred)
```

```python
gsc_model_02 = GridSearchCV(dia_model02, param_grid=params, cv=10 , refit=True ,scoring='recall')
gsc_model_02.fit(strain_data , y_train)
```
