```python
# target 값 정수화 함수 

def feature_encoder(df) :
    features = ['label_역률평균' , 'label_전류고조파평균' , 'label_전압고조파평균']
    
    for f in features :
        le = LabelEncoder()
        le = le.fit(df[f])
        df[f] = le.transform(df[f])
    
    return df 
```
